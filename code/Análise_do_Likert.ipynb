## **Lendo o datasett**

```python
url = 'https://docs.google.com/spreadsheets/d/1C2dAnRKa8BEngfjgHopLXYnUWoG2K_LW9szLuC3ABLM/edit?usp=sharing'
file_id = url.split('/')[-2]
read_url=f'https://docs.google.com/spreadsheets/d/{file_id}/export?format=xlsx'

# read the data
df = pd.read_excel(read_url, sheet_name='tabela1')
df
```

```python
another_scale = ['Discordo Totalmente', 'Discordo Parcialmente', 'Neutro / Indiferente', 'Concordo Parcialmente', 'Concordo Totalmente']
```

```python
df.replace({1: 'Discordo Totalmente', 2: 'Discordo Parcialmente', 3: 'Neutro / Indiferente', 4: 'Concordo Parcialmente', 5: 'Concordo Totalmente' },  inplace=True )
```

```python
df
```

```python
plot_likert.plot_likert(df, another_scale,  figsize=(5,10), bar_labels=True, bar_labels_color="snow", colors=plot_likert.colors.default_with_darker_neutral);
```
